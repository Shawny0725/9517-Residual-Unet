```python
import cv2
from matplotlib import pyplot as plt
import skimage.io as io
import numpy as np
```

```python
img = io.imread("train-labels00.jpg",as_gray= True)
print(img)
```

```python
print(img.shape)
print(np.max(img))
```

```python

# img=(img>125)+0
# img = img / 255
img[img > 128] = 255
img[img <= 128] = 0

plt.imshow(img,'gray')
plt.show()
```

```python
io.imsave('label.jpg',img,quality=100)
```

```python
from tifffile import imsave
labletif = io.imread("train-labels00.jpg",as_gray= True)
labletif = labletif.astype(np.float32)
print(np.unique(labletif))
imsave('label.tif',labletif)
```
